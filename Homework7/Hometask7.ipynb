```python
import requests

def fetch_posts():
    # URL API для получения постов
    url = "https://jsonplaceholder.typicode.com/posts"

    try:
        # Отправляем GET-запрос
        response = requests.get(url)

        # Проверяем, успешно ли прошел запрос (код 200)
        response.raise_for_status()

        # Преобразуем ответ в формат JSON (список словарей)
        posts = response.json()

        # Извлекаем и выводим первые 5 постов
        print(f"--- Первые 5 постов из {len(posts)} ---\n")
        
        for i, post in enumerate(posts[:5], 1):
            title = post.get('title')
            body = post.get('body').replace('\n', ' ') # Убираем переносы строк для красоты
            
            print(f"Пост №{i}")
            print(f"Заголовок: {title}")
            print(f"Тело: {body}")
            print("-" * 30)

    except requests.exceptions.RequestException as e:
        print(f"Произошла ошибка при запросе: {e}")

if __name__ == "__main__":
    fetch_posts()
```
